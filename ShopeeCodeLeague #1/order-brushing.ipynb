```python
df = pd.read_csv('../input/shopee/df.csv')
```

```python
df
```

```python
df = df.drop('orderid', axis=1)
```

```python
df = df.sort_values(by=['shopid', 'event_time'])
df
```

```python
data = df.groupby(['shopid'], 
                  as_index=False)['event_time', 'userid'].agg(lambda x: list(x))
```

```python
data['Aneh']=''
data['Orang']=''
```

```python
data
```

```python
for x in range(len(data['shopid'])):
    print(x, data['shopid'][x])
    maks=0
    for y in range(len(data['event_time'][x])):
        daftar=[]
        aneh=''
        tot=0
        for z in range(y,len(data['event_time'][x])):
            if (data['event_time'][x][z]-data['event_time'][x][y])<3600:
                daftar.append(data['userid'][x][z])
                tot+=1
            else:
                break
        daftarunik = set(daftar)
        panjangunik = len(daftarunik)
        if tot>=maks:
            maks=tot
            if (tot/panjangunik)>=3:
                if data['Aneh'][x]=='True':
                    list(data['Orang'][x]).append(daftar)
                else:
                    aneh='True'
                    data['Aneh'][x]=aneh
                    data['Orang'][x]=daftar
        else:
            pass
```
