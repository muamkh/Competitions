```python
import pandas as pd
import numpy as np
import time
from datetime import datetime,date,timedelta
```

```python
df = pd.read_csv('../input/logistics-shopee-code-league/delivery_orders_march.csv')
```

```python
df
```

```python
sub = pd.DataFrame(data=df['orderid'],columns=['orderid'])
sub['is_late']=0
sub
```

```python
for x in range(len(df['orderid'])):
    print(x)
    tot=0
    if pd.isna(df['2nd_deliver_attempt'][x]):
        fromdate = datetime.fromtimestamp(df['pick'][x]).strftime("%Y-%m-%d")
        if datetime.fromtimestamp(df['pick'][x]).weekday()==6:
            fromdate-=1
        todate = datetime.fromtimestamp(df['1st_deliver_attempt'][x]).strftime("%Y-%m-%d")
        tot = np.busday_count(fromdate, todate, weekmask='1111110',holidays=['2020-03-25','2020-03-30','2020-03-31'])
        if df['selleraddress'][x][-1]=='a':
            if (df['buyeraddress'][x][-1]=='o' or df['buyeraddress'][x][-1]=='s') and tot>7:
                sub['is_late'][x]=1
            elif df['buyeraddress'][x][-1]=='n' and tot>5:
                sub['is_late'][x]=1
            elif df['buyeraddress'][x][-1]=='a' and tot>3:
                sub['is_late'][x]=1
            else:
                sub['is_late'][x]=0
        elif df['selleraddress'][x][-1]=='n':
            if (df['buyeraddress'][x][-1]=='o' or df['buyeraddress'][x][-1]=='s') and tot>7:
                sub['is_late'][x]=1
            elif (df['buyeraddress'][x][-1]=='n' or df['buyeraddress'][x][-1]=='a') and tot>5:
                sub['is_late'][x]=1
            else:
                sub['is_late'][x]=1
        elif (df['selleraddress'][x][-1]=='s' or df['selleraddress'][x][-1]=='o'):
            if tot>7:
                sub['is_late'][x]=1
            else:
                sub['is_late'][x]=0
    else:
        fromdate = datetime.fromtimestamp(df['1st_deliver_attempt'][x]).strftime("%Y-%m-%d")
        todate = datetime.fromtimestamp(df['2nd_deliver_attempt'][x]).strftime("%Y-%m-%d")
        tot = np.busday_count(fromdate, todate, weekmask='1111110',holidays=['2020-03-25','2020-03-30','2020-03-31'])
        if tot>3:
            sub['is_late'][x]=1
        else:
            fromdate = datetime.fromtimestamp(df['pick'][x]).strftime("%Y-%m-%d")
            todate = datetime.fromtimestamp(df['1st_deliver_attempt'][x]).strftime("%Y-%m-%d")
            tot = np.busday_count(fromdate, todate, weekmask='1111110',holidays=['2020-03-25','2020-03-30','2020-03-31'])
            if df['selleraddress'][x][-1]=='a':
                if (df['buyeraddress'][x][-1]=='o' or df['buyeraddress'][x][-1]=='s') and tot>7:
                    sub['is_late'][x]=1
                elif df['buyeraddress'][x][-1]=='n' and tot>5:
                    sub['is_late'][x]=1
                elif df['buyeraddress'][x][-1]=='a' and tot>3:
                    sub['is_late'][x]=1
                else:
                    sub['is_late'][x]=0
            elif df['selleraddress'][x][-1]=='n':
                if (df['buyeraddress'][x][-1]=='o' or df['buyeraddress'][x][-1]=='s') and tot>7:
                    sub['is_late'][x]=1
                elif (df['buyeraddress'][x][-1]=='n' or df['buyeraddress'][x][-1]=='a') and tot>5:
                    sub['is_late'][x]=1
                else:
                    sub['is_late'][x]=1
            elif (df['selleraddress'][x][-1]=='s' or df['selleraddress'][x][-1]=='o'):
                if tot>7:
                    sub['is_late'][x]=1
                else:
                    sub['is_late'][x]=0
```
